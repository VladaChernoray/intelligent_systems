Варіант №3\
\begin{cases}
8.2 x_1 + 7.9 x_2 = 123 \\
11.2 x_1 + 16.3 x_2 = 201 \\
-16.3 x_1 - 8.3 x_2 = 169
\end{cases}

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

eq1 = lambda x: (123 - 8.2*x) / 7.9
eq2 = lambda x: (201 - 11.2*x) / 16.3
eq3 = lambda x: (-169 + 16.3*x) / (-8.3)

x = np.linspace(-50, 50, 100)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12,4))

ax1.plot(x, eq1(x), linestyle='--', color='b', label='')
ax1.set_title('8.2 x1 + 7.9 x2 = 123')
ax1.set_ylim([-20, 20])

ax2.plot(x, eq2(x), linestyle='-.', color='orange', label='')
ax2.set_title('11.2 x1 + 16.3 x2 = 201')
ax2.set_ylim([-20, 20])

ax3.plot(x, eq3(x), linestyle=':', color='g', label='')
ax3.set_title('–16.3 x1 – 8.3 x2 = 169')
ax3.set_ylim([-20, 20])

fig.suptitle('Graphs of Three Equations', fontsize=14)

plt.show()
```

```python
# import matplotlib.pyplot as plt
import numpy as np

eq1 = lambda x: (123 - 8.2*x) / 7.9
eq2 = lambda x: (201 - 11.2*x) / 16.3
eq3 = lambda x: (-169 + 16.3*x) / (-8.3)

x = np.linspace(0, 15, 100)

line_styles = ['--', '-.', ':']
line_colors = ['violet', 'b', 'g']

fig, ax = plt.subplots(figsize=(7, 7))

ax.plot(x, eq1(x), linestyle=line_styles[0], color=line_colors[0], label='8.2 x1 + 7.9 x2 = 123')
ax.plot(x, eq2(x), linestyle=line_styles[1], color=line_colors[1], label='11.2 x1 + 16.3 x2 = 201')
ax.plot(x, eq3(x), linestyle=line_styles[2], color=line_colors[2], label='–16.3 x1 – 8.3 x2 = 169')

ax.set_xlim(0, 15)
ax.set_xticks(np.arange(0, 15, 1))
ax.set_ylim(0, 15)
ax.set_yticks(np.arange(0, 15, 1))

ax.grid(True)
ax.legend()

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('All lines in one figure')

ax.text(1.3, 11.2, '8.2 x1 + 7.9 x2 = 123', rotation=-46, color=line_colors[0])
ax.text(10, 2.9, '11.2 x1 + 16.3 x2 = 201',rotation=-35, color=line_colors[1])
ax.text(7.3, 2, '–16.3 x1 – 8.3 x2 = 169',rotation=-63, color=line_colors[2])

fill_x = np.linspace(0, 15, 100)
fill_y1 = eq1(fill_x)
fill_y2 = eq2(fill_x)
fill_y3 = eq3(fill_x)
ax.fill_between(fill_x, fill_y1, fill_y2, where=(fill_y2 <= fill_y1) & (fill_y2 >= fill_y3 - 0.1), color='lightGray')
ax.fill_between(fill_x, fill_y1, fill_y3, where=(fill_y3 <= fill_y1) & (fill_y3 >= fill_y2), color='lightGray')

points = []
for i, eqi in enumerate([eq1, eq2, eq3]):
    for j, eqj in enumerate([eq1, eq2, eq3]):
        if i < j:
            A = np.array([[8.2, 7.9], [11.2, 16.3], [-16.3, -8.3]])
            b = np.array([123, 201, -169])
            x = np.linalg.solve(A[[i, j], :], b[[i, j]])
            points.append(x)
            ax.plot(x[0], x[1], marker='.', markersize=5, color='r')
            ax.annotate('({:.1f}, {:.1f})'.format(x[0], x[1]), xy=x, xytext=(x[0] + 0.2, x[1] + 0.2), color='r')

fig.savefig('graph/three_equations.png', dpi=100)
fig.savefig('graph/three_equations.svg', dpi=100)
fig.savefig('graph/three_equations.jpg', dpi=100)
```
